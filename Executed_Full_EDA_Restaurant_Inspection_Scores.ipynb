# **Exploratory Data Analysis (EDA) - Restaurant Inspection Scores**

## **1. Load Libraries and Dataset**

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "ARCHIVED__Restaurant_Inspection_Scores__2016-2019__20250309.csv"
df = pd.read_csv(file_path)

df.info()
df.head()
```

## **2. Data Overview and Cleaning**

```python

num_cols = ['inspection_score', 'business_latitude', 'business_longitude', 
            'Neighborhoods', 'Current Police Districts', 
            'Current Supervisor Districts', 'Analysis Neighborhoods']

cat_cols = ['inspection_type', 'risk_category']

df_num = df[num_cols].dropna()

summary_stats = df[num_cols].describe()

summary_stats
```

## **3. One-Dimensional Analysis**

### **Histograms for Numerical Attributes**

```python

sns.set_style("whitegrid")

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

### **Bar Charts for Categorical Attributes**

```python

for col in cat_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette="viridis")
    plt.title(f"Bar Chart of {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.show()
```

## **4. Summary Statistics and Outliers**

```python

outlier_dict = {}
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))][col]
    outlier_dict[col] = len(outliers)

outlier_dict
```

## **5. Box Plot for Numerical Variables**

```python

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[num_cols])
plt.title("Box Plot of Numerical Attributes")
plt.xticks(rotation=45)
plt.show()
```

## **6. Pairwise Relationships**

### **Scatter Plot Matrix for Numerical Variables**

```python

sns.pairplot(df[num_cols].dropna())
plt.show()
```

### **Heatmap for Categorical vs Categorical Data**

```python

if 'inspection_type' in df.columns and 'risk_category' in df.columns:
    cross_tab = pd.crosstab(df['inspection_type'], df['risk_category'])
    plt.figure(figsize=(10, 6))
    sns.heatmap(cross_tab, annot=True, fmt="d", cmap="coolwarm")
    plt.title("Heatmap of Inspection Type vs Risk Category")
    plt.show()
```

### **Box Plots for Categorical vs Numerical Data**

```python

for num_col in ['inspection_score']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df['risk_category'], y=df[num_col], data=df)
    plt.title(f"Risk Category vs {num_col}")
    plt.xticks(rotation=45)
    plt.show()
```

### **Categorical vs Numerical: Using Color to Show Relationships**

```python

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['inspection_score'], y=df['business_latitude'], hue=df['risk_category'], palette='viridis', alpha=0.7)
plt.title("Inspection Score vs Business Latitude (Colored by Risk Category)")
plt.xlabel("Inspection Score")
plt.ylabel("Business Latitude")
plt.show()
```
